```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_excel('한국성씨.xlsx')
plt.rc('font', family = 'Malgun Gothic')
plt.title('한국 성씨 Top7') # 차트 제목 넣기
plt.axis('equal')

colors_list = ['#2743D2', '#0066ff', '#3385ff', '#4d94ff', '#66a3ff', '#99c2ff', '#e6f0ff'] # red 가능

wedgeprops = {'width':0.6, 'edgecolor':'black', 'linewidth':2} #부채꼴 영역의 스타일을 설정

plt.pie(data['2015년'], labels = data['성씨'], 
        colors = colors_list, # 색상 넣기
       startangle = 90, counterclock = False, #자료의 첫 시작 각도와 시계방향으로 설정하기
       autopct = '%.1lf%%', # 표시될 숫자의 형식을 지정
       shadow = True, #그림자 
       wedgeprops = wedgeprops) #부채꼴 모양

plt.legend(loc = 2) #범례
plt.show()
```

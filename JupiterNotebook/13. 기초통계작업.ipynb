## 년월로 그룹화하여 전화 평균 구하기

```python
print(df.groupby('년월')['전화'].mean()) #2022 1월과 2022년 2월로 나누어 평균 구함
```

```python
print(df.groupby('년월')['전화'].count()) # 각각 15건 있었다는 뜻
```

## 년도별 방문 횟수

```python
print(df.groupby('년월')['방문'].sum())
```

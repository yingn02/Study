## Dataframe 객체 생성

```python
data = {'종가':[71200, 69500, 70100, 71500, 72900, 71700],
       '전일비':['▲1700','▼600','▼1400','▼1400','▲1200','▼400'],
       '등락률(%)':[2.45, -0.86, -1.96, -1.92, 1.67, -0.55],
       '시가':[70800, 68800, 70000, 72700, 72300, 72300],
       '고가':[71200, 7000, 70600, 72700, 73100, 72400],
       '저가':[70500, 68700, 69900, 71200, 72200, 71500],
       '거래량':[21354798, 15828269, 18617138, 13409634, 13232638, 12481430]}
data
```

```python
import pandas as pd
df = pd.DataFrame(data)
df
```

## Index 만들기

```python
df = pd.DataFrame(data, 
                index = ['2022-03-10', '2022-03-08', '2022-03-07', '2022-03-04', '2022-03-03', '2022-03-02'])
df
```

## DataFrame 객체 생성 (Columns 지정)

### data중에서 원하는 column만 선택하거나, 순서 변경 지정

```python
df = pd.DataFrame(data,
                 index = ['2022-03-10', '2022-03-08', '2022-03-07', '2022-03-04', '2022-03-03', '2022-03-02'],
                 columns = ['종가', '거래량']) #종가와 거래량만 표시되게 한다
df
```

```python
df = pd.DataFrame(data,
                 index = ['2022-03-10', '2022-03-08', '2022-03-07', '2022-03-04', '2022-03-03', '2022-03-02'],
                 columns = ['종가', '거래량', '전일비'])  #종가와 거래량과 전일비만 표시되게 한다
df
```

## 날짜 입력받고 해당 날짜의 주식 시세 알아보기

```python
date = input("삼성주가 날짜 입력 => ") #날짜 입력받기
a = df.index.str.contains(date) #데이터 프레임의 인덱스 문자열에 원하는 문자열이 포함된 행을 찾아냄
df[a]
```

```python
# 위에 종가, 거래량, 전일비만 나오는 이유는 columns가 아직 적용되어있기 때문이다.
#이것을 해제하려면 columns가 없는 상태로 다시 선언을 해주어야한다

df = pd.DataFrame(data,
                 index = ['2022-03-10', '2022-03-08', '2022-03-07', '2022-03-04', '2022-03-03', '2022-03-02'])
                
df
```

```python
date = input("삼성주가 날짜 입력 => ") #날짜 입력받기
a = df.index.str.contains(date) #데이터 프레임의 인덱스 문자열에 원하는 문자열이 포함된 행을 찾아냄
df[a]
```

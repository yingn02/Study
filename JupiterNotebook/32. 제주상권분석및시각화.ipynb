```python
# 4. 마커 표시하기
# location: 마커를 추가할 위도/경도 좌표를 입력 후
# popup: 표기할 팝업 문구 지정 (마우스 클릭시 표기)
# tooltip: 표기할 툴팁 지정 (마우스 오버시 표기)
import folium

map =  folium.Map(location=[33.5137542483577,126.5368466], zoom_start = 15)

folium.Marker(location=[33.5137542483577,126.5368466],
              popup="세흥장식", # popup 은 마커를 클릭하면 뜸 tooltip 은 마커에 마우스를 올리면 뜸 icon은 마커 모양 정함
              tooltip="제주특별자치도 제주시 동문로",
             icon = folium.Icon('green', icon = 'star')).add_to(map)  

map
```

```python
# 7. 제주 상권 정보 파일 가져오기
import pandas as pd
import warnings                    # 파이썬 코딩시 warnings 무시
warnings.filterwarnings('ignore')  # 파이썬 코딩시 warnings 무시

df = pd.read_csv('소상공인시장진흥공단_상가(상권)정보_제주_202203.csv')
df.head(5)
```

```python
# 8. 제주 상권 정보 파일에서 컬럼명 살펴보기
pd.Series(df.columns) #필드명들을 볼 수 있다
```

```python
# 9. 필요한 컬럼 정보만 가져오기

jeju = pd.read_csv('소상공인시장진흥공단_상가(상권)정보_제주_202203.csv')

# 필요한 컬럼 정보만 가져옴
jeju_sub = jeju[['시군구명', '상권업종대분류명', '상권업종중분류명', '위도', '경도','도로명','행정동명']]
jeju_sub.head(10)
```

```python
# 10. isin() DataFrame이나 Series에 존재 하는지를 True(존재), False(존재안함)로 반환
sub_df = df.loc[df['행정동명'].isin(['건입동','일도이동'])] # isin(), 특정 지역만 가져옴
sub_df.head(5)
```

```python
# 11. 행정동명이 건입동, 일도이동 행정명 위도, 경도
sub_df[['위도', '경도', '상호명','행정동명']]   #행정동명이 건입동, 일도이동 행정명 위도, 경도
```

```python
# 12. MarkerCluster
# MarkerCluster 를 이용하면 가까운 거리의 Marker들을 군집시켜서 해당 건수를 표현해준다.
# 특정 클러스터를 클릭하면 해당 위치를 확대시켜 자세히 들여다볼 수 있다.

m = folium.Map(location=[33.5137542483577,126.5368466]) # 지도 만들고

folium.CircleMarker([33.5137542483577,126.5368466], 
                    color='tomato',
                     radius = 50, 
                    tooltip='제주 상권').add_to(m) #원을 그림
m
```

```python
# 13. 가까운 거리의 Marker들을 군집시켜서 해당 건수를 표현

from folium.plugins import MarkerCluster


m = folium.Map(location=[33.5137542483577,126.5368466],zoom_start=15)

coords = sub_df[['위도', '경도']]

marker_cluster = MarkerCluster().add_to(m)

for lat, long in zip(coords['위도'], coords['경도']): # zip, 리스트들을 튜플형태로 묶어줌
    folium.Marker([lat, long], icon = folium.Icon(color="green")).add_to(marker_cluster)
m
```

```python
# 14. 행정동명이 건입동, 일도이동 지역의 상권에서 커피점/카페 검색

jeju_coffee = sub_df.loc[sub_df['상권업종중분류명'] == '커피점/카페']
jeju_coffee =jeju_coffee[['위도', '경도', '상호명','행정동명','상권업종중분류명']]   #행정동명이 건입동, 일도이동 행정명 위도, 경도
jeju_coffee
```

```python
# 15. 행정동명이 건입동, 일도이동 지역의 상권에서 커피점/카페 지도에 표시

from folium.plugins import MarkerCluster


m = folium.Map(location=[33.5137542483577,126.5368466],zoom_start=15)

coords = jeju_coffee[['위도', '경도']]

marker_cluster = MarkerCluster().add_to(m)

for lat, long in zip(coords['위도'], coords['경도']):   #zip()는 반복문 수행시 두개의 리스트 값을 묶어준다.
    folium.Marker([lat, long], icon = folium.Icon(color="green")).add_to(marker_cluster)
m
```

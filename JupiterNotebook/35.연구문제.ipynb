## 3. 데이터 분석 : 연구문제 관련 데이터 분석 (1), (2)

```python
# 데이터 분석하기(pandas사용)



### (1) 경기도 구리시의 상권 업종분류에 대해 알아보기

import pandas as pd # pandas로 불러오기

df = pd.read_csv('구리시_상권정보.csv', encoding = 'cp949') #csv 파일 불러오기
df2 = df['상권업종대분류명']
df2
```

```python
### (2) 경기도 구리시에 있는 상권 업종의 분포 순위를 알아보기

import pandas as pd # pandas로 불러오기
from collections import Counter

df3 = pd.read_csv('구리시_상권정보.csv', encoding = 'cp949', index_col = 0) #csv 파일 불러오기

s_list = list(df3['상권업종대분류명']) # 리스트에 '상권업종대분류명'에 해당하는 자료들을 삽입

count = Counter(s_list)
num = count.most_common(50) #상권업종의 종류 n가지 중에 50종류까지 저장 허용
print(num) #저장한 것 출력
```

## 4. 시각화하기 : 연구문제 관련 데이터 시각화 (3), (4), (5)

```python
# 지도활용 데이터 시각화하기(foliun패키지 사용)

import folium
from folium.plugins import MarkerCluster #folium패키지 사용

### (3) 경기도 구리시 수택동에 가장 많이 분포한 상권 업종이 주로 어디 지역에 분포되어있는지 알아보기
#  위에서 가장 많이 분포한 상권 업종은 소매(2797 곳) 이었으므로, '소매'로 분류된 상권을 모두 지도에 표시한다


sub_df = df.loc[df['법정동명'].isin(['수택동'])] # isin(), 특정 지역 (수택동) 만 가져옴
sub_df.head(5)

# 법정동명이 '수택동'인 지역의 상권에서 '소매' 검색

guri_s = sub_df.loc[sub_df['상권업종대분류명'] == '소매']
guri_s = guri_s[['위도', '경도', '상호명','법정동명','상권업종대분류명']]

# 법정동명이 '수택동'인 지역의 상권에서 '소매' 지도에 표시

map1 = folium.Map(location=[37.59201410, 127.1427998], zoom_start=15) # 수택동의 위도 경도를 입력하고, 15만큼 확대하여 지도로 나타냄

# 어디에 분포하는지 그리기
coords = guri_s[['위도', '경도']]
marker_cluster = MarkerCluster().add_to(map1)

#zip()을 이용하여 반복문 수행시 두개의 리스트 값을 묶어준다
for lat, long in zip(coords['위도'], coords['경도']): 
    folium.Marker([lat, long], icon = folium.Icon(color="green")).add_to(marker_cluster)
    
map1 #출력
```

```python
### (4) 위의 지도에서 병원과 가장 가까이 있는 곳이 어딘지 찾고 마커 표시하기
# 검배로 부근에 14곳이 밀집되어 있는 지역에서, 교문외과, 구산한의원이라는 병원에서 가장 가까운 '소매' 상권 발견함.

map2 =  folium.Map(location=[37.59662225,127.1449633], zoom_start = 20) # 위도 경도를 입력하고, 20만큼 확대하여 지도로 나타냄

folium.Marker(location=[37.59662225,127.1449633],
              popup="구산한의원 부근", # popup 은 마커를 클릭하면 뜸 tooltip 은 마커에 마우스를 올리면 뜸 icon은 마커 모양 정함
              tooltip="병원과 가장 가까운 '소매' 상권",
             icon = folium.Icon('green', icon = 'star')).add_to(map2)  

map2
```

```python
### (5) 상호명을 검색하면 경기도 구리시 수택동에 있는 상권들을 표시하도록 해보기

sub_df = df.loc[df['법정동명'].isin(['수택동'])] # isin(), 특정 지역 (수택동) 만 가져옴
sub_df.head(5)

# 법정동명이 '수택동'인 지역의 상권에서 상호명색

search = input("검색할 상호명을 입력하세요>> ") # 검색어 입력받기

guri_s = sub_df.loc[sub_df['상호명'] == search] # 검색어와 일치하는 것 찾기
guri_s = guri_s[['위도', '경도', '상호명','법정동명','상호명']]

# 법정동명이 '수택동'인 지역의 상권에서 검색한 상권을 지도에 표시

map3 = folium.Map(location=[37.59201410, 127.1427998], zoom_start=15) # 수택동의 위도 경도를 입력하고, 15만큼 확대하여 지도로 나타냄

# 어디에 분포하는지 그리기
coords = guri_s[['위도', '경도']]
marker_cluster = MarkerCluster().add_to(map3)

#zip()을 이용하여 반복문 수행시 두개의 리스트 값을 묶어준다
for lat, long in zip(coords['위도'], coords['경도']): 
    folium.Marker([lat, long], icon = folium.Icon(color="green")).add_to(marker_cluster)
    
map3 #출력
```

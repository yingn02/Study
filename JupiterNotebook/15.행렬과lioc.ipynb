## iloc  (i는 index 라는 의미)

```python
df =  pd.read_csv('samsung2.csv', encoding = 'cp949')
df
```

```python
print(df.iloc[24]) 
```

```python
print(df.iloc[-1]) # 파이썬 처럼 -1 가능, 맨 마지막 것 가져옴
```

# 데이터 추출할 때 - 행과 열을 (둘 다 사용)

```python
subset = df.loc[:, ['종가', '전일비']] #모든 행(:) 그리고 종가,전일비를 가져옴
subset
```

```python
#개인 실습 (10행, 열은 종가 전일비 거래량) loc
subset2 = df.loc[0:9, ['종가', '전일비', '거래량']]
subset2
```

```python
#개인 실습2 (10행, 열은 종가 전일비 거래량) iloc
subset3 = df.iloc[0:10, [1, 4, -1]]
subset3
```

```python
s_range = list(range(0,101,20)) #파이썬의 range와 유사, 0부터 100까지 20씩 증가하여 리스트 대입
s_range
```

```python
s_range = list(range(0,7,2))
s_range
subset = df.iloc[:, s_range] #열이 s_range, 0 2 4 6, range(0,8,1) 이면 모두 출력이 될 것
subset
```

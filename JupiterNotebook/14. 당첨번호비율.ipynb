## 로또 복권 빈도수

```python
import pandas as pd
from collections import Counter

df = pd.read_excel('lotto.xlsx', index_col = 0)

num_list = list(df['당첨번호1'])
num_list += list(df['당첨번호2']) #int 아니었다면 list(df['당첨번호2'].astype(int))
num_list += list(df['당첨번호3']) # num_list에 리스트를 합치는 작업
num_list += list(df['당첨번호4'])
num_list += list(df['당첨번호5'])
num_list += list(df['당첨번호6'])

count = Counter(num_list)
most_num = count.most_common(45)
most_num.sort() # 정렬, 안하면 많이 나온 순서대로 출력된다
print(most_num)
```

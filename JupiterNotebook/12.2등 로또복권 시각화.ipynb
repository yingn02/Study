```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('lotto.xlsx')
a = df['2등당첨금액']

def to_int(a):
    a = a.replace(',', '')
    a = a.replace('원', '')
    a = int(a)
    return a

a_int = df['2등당첨금액'].apply(to_int)
df['2등당첨금액'] = a_int

plt.rc('font', family = 'Malgun Gothic')

plt.title('로또복권 2등 당첨금액') #제목
plt.xlabel('회차') #x축 레이블
plt.ylabel('당첨금액(단위:억원)') #y축 레이블

x = df['회차'].iloc[:15].values #10까지
y = df['2등당첨금액'].iloc[:15].values / 100000000
color_list = ['orange', 'gray'] * 10

plt.bar(x, y, width = 0.5, color = color_list, edgecolor = 'lightgray', linewidth = 3)
plt.show()
```

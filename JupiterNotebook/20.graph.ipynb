### matplotlib은 다양한 형태의 그래프를 통해서 시각화 할 수 있는 라이브러리

### 1. 그래프 기본 그리기

```python
import matplotlib.pyplot as plt
plt.plot([10,20,30,40]) #입력 (y축)
plt.show() #출력
```

```python
x = [1,2,3,4,5]
y = [2,4,6,10,15]
plt.plot(x, y) # plot() 함수에 x축, y축 데이터 입력
plt.show() # show() 그래프를 보여줌
```

### 2. 그래프에 옵션 추가하기

### 그래프에 x축, y축 레이블 넣기

```python
plt.xlabel('X-Axis') # X축에 이름 붙여줌
plt.ylabel('Y-Axis') # Y축에 이름 붙여줌
plt.plot([1,2,3,4,5], [2,4,6,10,15]) #다른방식으로 대입
plt.show()
```

### 그래프에 제목 넣기

```python
plt.title('제목') #한글넣으면 깨짐
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.plot([1,2,3,4,5], [2,4,6,10,15])
plt.show()
```

```python
plt.rc('font', family = 'Malgun Gothic') #해결방법 (맑은 고딕)
plt.title('제목')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.plot([1,2,3,4,5], [2,4,6,10,15])
plt.show()
```

### 그래프에 범례 넣기
범례란 각 계열릐 이름을 표시하기 위해 차트의 한 공간에 표시된 차트의 요소

```python
plt.rc('font', family = 'Malgun Gothic')
plt.title('제목')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.plot([1,2,3,4,5], [2,4,6,10,15], label = 'Price($)') # 범례
plt.legend() # 범례
plt.show()
```

```python
plt.rc('font', family = 'Malgun Gothic')
plt.title('제목')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.plot([1,2,3,4,5], [2,4,6,10,15], label = 'Price($)') # 범례
plt.legend(loc = (0, 0)) # 범례 표시 위치도 설정 가능
plt.show()
```

### 그래프에 선 색 넣기

```python
plt.rc('font', family = 'Malgun Gothic')
plt.title('제목')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.plot([1,2,3,4,5], [2,4,6,10,15], label = 'Price($)', color = 'green') # 색 넣기
plt.legend(loc = (0, 0))
plt.show()
```

### 그래프 선 종류 지정하기

```python
plt.rc('font', family = 'Malgun Gothic')
plt.title('제목')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.plot([1,2,3,4,5], [2,4,6,10,15], label = 'Price($)', color = 'green', linestyle = '--') # 종류 지정
plt.legend(loc = (0, 0))
plt.show()
# - -- -. : 등 여러가지 있다, linestyle 대신 ls 가능
```

### 그래프 마커 모양 변경하기

```python
plt.rc('font', family = 'Malgun Gothic')
plt.title('제목')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.plot([1,2,3,4,5], [2,4,6,10,15], 'D', #데이터 바로 뒤에 씀
         label = 'Price($)', color = 'green', linestyle = '--') #color없애고 D를 rd로 바꾸면 빨강됨
plt.legend(loc = (0, 0))
plt.show()
```

### Bar 차트 그리기

```python
import matplotlib.pyplot as plt

y = [5,3,7,19,9]
x = range(len(y)) #range(5) 0 ~ 4

plt.bar(x, y, width = 0.7, color = ['r', 'g', 'b', 'y', 'c'])
plt.show()
```

```python
import matplotlib.pyplot as plt

y = [5,3,7,19,9]
x = range(len(y))

plt.barh(x, y, color = ['r', 'g', 'b', 'y', 'c']) # 가로로
plt.show()
```

### PI차트 그리기

```python
import matplotlib.pyplot as plt

ratio = [20,30,10,40]
labels = ['사과', '바나나', '배', '포도']

plt.pie(ratio, labels = labels, autopct = '%.2lf%%') #autopct : 각 파이 문자열 출력 형식 설정
plt.show()
```

```python
import matplotlib.pyplot as plt

ratio = [20,30,10,40]
labels = ['사과', '바나나', '배', '포도']

plt.title('과일 인기도')

plt.pie(ratio, labels = labels, autopct = '%.2lf%%', 
        explode = (0.1, 0.1, 0.1, 0.1), 
        colors = (['r', 'g', 'b', 'c']))

plt.show()
```

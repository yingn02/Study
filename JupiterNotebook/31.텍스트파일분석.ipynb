```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

f = open('test.txt','rt', encoding='UTF-8').read() #텍스트 파일 열어주기

wordcloud = WordCloud(font_path = 'C:/Windows/Fonts/malgun.ttf', #한글폰트 반드시 지정하기
        background_color="black", #배경
        width = 1000, 
        height = 1000,
        max_words = 1000, # 최대 단어 수
        max_font_size = 10000) # 최대 폰트 사이즈

wordcloud = wordcloud.generate(f) # f 대입 

plt.imshow(wordcloud)
plt.axis('off') #축 가리기                   
plt.show()   
```

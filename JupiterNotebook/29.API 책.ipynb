## 네이버 책에서 정보 가져오기

```python
# 네이버
import os
import sys
import urllib.request
import pandas as pd
import json
import re

client_id = "7MJsiwZO5aIph9spMJg6"
client_secret = "qaEkrVmVIH"

query = urllib.parse.quote(input("검색질의: ")) #검색어 입력 받아서 진짜로 검색하는거임
idx = 0
display = 100 #최대 100건까지 검색결과 보여줌
start = 1 #검색 시작 위치 1
end = 1000
sort = "sim"

book_df = pd.DataFrame(columns = ('Title', 'Link', 'Image', 'Author', 'Price', 'Discount', 'Publisher', 'description', 'Pubdate'))

for start_index in range(start, end, display):
    # json 결과, 뉴스 겁색
    url = "https://openapi.naver.com/v1/search/book?query=" + query \
    + "&display=" + str(display)\
    + "&start=" + str(start_index)\
    + "&sort=" + sort #sort는 문자열이라 굳이 바꿀 필요 X
    
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    
    rescode = response.getcode()
    if(rescode==200): #정상적으로 응답받음
        response_body = response.read()
        response_dic = json.loads(response_body.decode('utf-8')) #json으로 들어온 객체를 딕셔너리로 받아오기
        print(response_dic)
        items = response_dic['items'] # item 키에 해당하는 값 가져오기
        
        for item_index in range(0,len(items)):
            # 태그를 제거하는 변수, <.*?> 는 태그안에 있는 모든 것들을 제거할 수 있게함
            r_tag = re.compile('<.*?>') #태그를 제외하기 위한 패턴 지정하기
            
            title = re.sub(r_tag, '', items[item_index]['title']) # 정규식과 매치되는 부분을 다른 문자로 변경할 수 있다
            link = items[item_index]['link']
            image = items[item_index]['image']
            author = items[item_index]['author']
            price = items[item_index]['price']
            discount = items[item_index]['discount']
            publisher = items[item_index]['publisher']
            description = items[item_index]['description']
            pubdate = items[item_index]['pubdate']
            
            book_df.loc[idx] = [title, link, image, author, price, discount, publisher, description, pubdate]
            idx += 1
            
    else:
        print("Error Code:" + rescode)
        
#book_df
book_df.to_excel('파이썬.xlsx') #엑셀로 저장해보자
```
